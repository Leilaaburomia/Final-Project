# Introduction

Over the past decade, the western United States has experienced hydroclimatic variability that has emerged as a dominant environmental issue of great concern (@Balling2007). Since 1996, a considerable portion of the western U.S. has seen drought conditions that are similar in scale to the historical droughts in the 1930s and 1950s (@Balling2007). Studies have also used the paleoclimatic record to reveal evidence of long term droughts over the past 500 to 2,000 years (@Woodhouse1998). Hydroclimatic variability studies range in scale from regional to watershed. This variability is often complex and may require a multitude of scales to fully understand and plan for future drought conditions. In the dry western U.S., Colorado receives less than 17 inches of average annual precipitation statewide. (@Mckee2000). In comparison to the rest of the United States, these rainfall amounts are incredibly sparse. Luckily, Colorado’s Rocky Mountains help alleviate potential water stress by collecting more than 25 inches of precipitation annually (@Mckee2000). By successfully adapting to this dry climate, Colorado supports a massive agricultural and recreational industry and multiple growing urban centers (@Mckee2000). As population and industry grow, the threat of drought and its social, economic, and environmental impacts increase as well. Measuring and assessing drought accurately is essential for future planning. Historical drought data can give valuable insight into influences or indicators of drought conditions. Patterns of climatic variability that aligned with droughts in the past can help our society better prepare for the future. One of these patterns is the influential El Nino and La Nina. El Nino is caused by a warming of the tropical Pacific Ocean that occurs about every three to seven years and lasts for 12-18 months (@McPhaden2002). La Nina is akin to El Nino with stronger than normal trade winds and cooler tropical Pacific water temperatures (@McPhaden2002). In recent decades, the El Nino/Southern Oscillation (ENSO) has been more closely monitored, and the strength of each cycle has been documented. The effects of El Nino and La Nina on climate vary considerably across the globe, but in the western U.S., there is a pattern of impacts. During El Nino years, both extremely wet and dry conditions can be expected, while La Nina years are characterized by only dry conditions (@Canon2007). This research aims to analyze interactions between climatic variability indicating drought and the ENSO cycle. This condition can be measured on a variety of time scales, indicating that the SPI can be useful for both short-term and long-term applications (@NASA). This data uses six classifications to characterize drought severity: normal conditions, abnormally dry (D0), and four levels of drought (D1 to D4) representing moderate, severe, extreme, and exceptional drought conditions, respectively. Lastly, we will utilize SNOTEL snow pack data from the Natural Resource Conservation Service to supplement our drought framework. In order to compare this data to ENSO cycles, we will examine El Niño and La Niña intensities for each year based on the Oceanic Nino Index (ONI). This index is the standard for NOAA used to classify El Nino and La Nina events in the eastern tropical Pacific (@Null). Events are defined as five consecutive 3-month periods at or above the + 0.5 degrees threshold for warm (El Nino) events and at or below the - 0.5 degrees anomaly for cool (La Nina) events (@Null). ### Time Series Plots

```
# more data cleaning

library(dplyr)

data1 <- data1 %>%
  rename(Date = `Date (for Reference)`)

long_data <- long_data %>%
  rename(Date = `Date_(for_Reference)`)

long_data <- long_data %>%
  left_join(data1 %>% select(Date, Rolling_3_Month_Date = `Rolling 3 Month Date`), by = "Date")
```

#### Create a “Center Date” for the Rolling 3-Month Period

```
# Create a lookup table for center months of each rolling period
period_centers <- c(
  "JFM" = "02", "FMA" = "03", "MAM" = "04", "AMJ" = "05",
  "MJJ" = "06", "JJA" = "07", "JAS" = "08", "ASO" = "09",
  "SON" = "10", "OND" = "11", "NDJ" = "12", "DJF" = "01"
)

# Add a "centered" date column
long_data <- long_data %>%
  mutate(
    center_month = period_centers[`Rolling_3_Month_Date`],
    year = lubridate::year(Date),
    rolling_date = as.Date(paste(year, center_month, "15", sep = "-"))
  )
```

#### Time Series Visualizations for SPI and ENSO

```
long_data <- long_data %>%
  mutate(
    ENSO_Phase = case_when(
      ENSO_Type %in% c("WE", "ME", "SE", "VSE") ~ "El Nino",
      ENSO_Type %in% c("WL", "ML", "SL") ~ "La Nina",
      ENSO_Type == "NONE" ~ "Neutral",
      TRUE ~ NA_character_ 
    )
  )

enso_bands <- long_data %>%
  select(rolling_date, ENSO_Phase) %>%
  distinct() %>%
  arrange(rolling_date) %>%
  mutate(group = cumsum(ENSO_Phase != lag(ENSO_Phase, default = first(ENSO_Phase)))) %>%
  group_by(group, ENSO_Phase) %>%
  summarise(start = min(rolling_date), end = max(rolling_date), .groups = "drop")

ggplot() +
  geom_rect(data = enso_bands,
            aes(xmin = start, xmax = end, ymin = -Inf, ymax = Inf, fill = ENSO_Phase),
            alpha = 0.2) +
  # SPI lines
  geom_line(data = long_data %>% filter(Metric_Type == "Rolling SPI Average"),
            aes(x = rolling_date, y = Value, color = County)) +
  scale_fill_manual(values = c("El Nino" = "red", "La Nina" = "blue", "Neutral" = "gray")) +
  labs(title = "SPI with ENSO Phase Overlays",
       x = "Date", y = "Rolling SPI Average",
       fill = "ENSO Phase", color = "County") +
  theme_minimal()
```

#### Time Series Visualizations for DSCI and ENSO

```
ggplot() +
  geom_rect(data = enso_bands,
            aes(xmin = start, xmax = end, ymin = -Inf, ymax = Inf, fill = ENSO_Phase),
            alpha = 0.2, inherit.aes = FALSE) +
  geom_line(data = long_data %>% filter(Metric_Type == "Rolling DSCI Average"),
            aes(x = rolling_date, y = Value, color = County)) +
  facet_wrap(~County, scales = "free_y") +
  scale_fill_manual(values = c("El Nino" = "red", "La Nina" = "blue", "Neutral" = "gray")) +
  labs(title = "DSCI Over Time with ENSO Phase Overlays",
       x = "Date", y = "Rolling DSCI Average",
       fill = "ENSO Phase", color = "County") +
  theme_minimal()
```

# Discussion

Looking at the visualization for the relationship between ENSO Severity and Rolling SPI Average, we can see that for our North, East, West, and Central counties (Larimer, Kit Carson, Mesa, and Park, respectfully), there is a negative weak to moderate correlation between ENSO severity and SPI. However our Southern County, Costilla, has a weak positive correlation between the two variables. When we run a linear regression on our five counties to identify if ENSO severity can predict SPI values, are results are as follows: 

-   1\. For Kit Carson County (East), we got a p-value of 0.0152 (significant), which suggests that ENSO drives drier conditions in this area, but with an R2 value of 0.02, there is weak explanatory power. 

-   2\. For Costilla County (South), we got a p-value of 0.174 (not significant), which suggests that SPI seems to increase with ENSO severity, but it’s not statistically significant. This could be due to noise or local variability. With an R2 value of 0.006, there is very weak explanatory power. 

-   3\. For Larimer County (North), we got a p-value of 0.0000000003 (very significant), which suggests a strong negative association. ENSO events have a marked drying effect in this county. With an R2 value of 0.14, there is moderate explanatory power. 

-   4\. For Mesa County (West), we got a p-value of 0.000025 (very significant), which suggests a strong negative association. Higher ENSO severity leads to lower SPI. With an R2 value of 0.058, there is low explanatory power, but meaningful. 

-   4\. For Park County (Central), we got a p-value of 0.000000018 (very significant), which suggests a strong negative association. Again, higher ENSO severity leads to lower SPI. With an R2 value of 0.10, there is low to moderate explanatory power.

We then ran the same statistical analyses (correlation test, and linear regression) to determine the relationship between ENSO severity and DSCI (drought) values. Before going through the results of our linear regression, it is important to look at our visualization of the data. We can see from the graph that Costilla and Mesa County have a slight positive correlation between ENSO severity and DSCI values, Larimer and Carson County have a slight negative correlation between the two variables, and Park county has practically no correlation. Here are the results of our linear regression:

-   1\. For Kit Carson County (East), we got a p-value of 0.0814 ( borderline significant), which suggests that there is a possible weak drought & ENSO connection here. Worth noting but not strong. An R2 value of 0.01, only 1% of DSCI variability is explained by ENSO.

-   2\. For Costilla County (South), we got a p-value of 0.769 (not significant at all), which suggests that ENSO doesn’t appear to affect DSCI in this county. With an R2 value of 0, there is essentially no relationship. 

-   3\. For Larimer County (North), we got a p-value of 0.231 (not significant), which suggests a very weak, nonsignificant relationship. ENSO events have little to no relationship to DSCI. With an R2 value of 0.005, this explains less than 1% of variability.

-   4\. For Mesa County (West), we got a p-value of 0.207 (not significant), which again suggests a very weak nonsignificant relationship. Higher ENSO severity has barely any effect on DSCI values. With an R2 value of 0.05, there is low explanatory power.

-   4\. For Park County (Central), we got a p-value of 0.932 (very far from significant), which suggests absolutely 0 relationship between ENSO and DSCI values. With an R2 value of 0, there is 0 explanatory power.

Our data implies that there is an overall negative association between ENSO severity and SPI values, and no significant relationship between ENSO severity and DSCI values. This goes against what we previously believed. We were under the assumption that a higher ENSO value (Stronger El Niño seasons), would result in higher precipitation values, leading to reduced drought conditions. Our initial assumption is true for the southern part of Colorado (Costilla County), but after looking at the statistical analyses, we can see that this correlation between both ENSO & SPI, and ENSO & DSCI is very weak, and not significant.

There also might be a shorter lag between ENSO severity and precipitation measurements. Other limits may include: small sampling size. Including more counties might get us a better picture. To clean the data, we had to average 3 or 4 monthly drought values to line up with our ENSO severity values, and SPI values. There may be data hidden in a more complex analysis.
